```python
import pandas as pd
import geopandas as gpd
```

```python
def extract_xy(df, xcoord='x', ycoord='y'):
    df[xcoord]=df['id'].str.slice(0,6).astype(int)/10000
    df[ycoord]=df['id'].str.slice(6,12).astype(int)/10000
    return df
```

```python
gdfperif=gpd.read_file(r'/mnt/nvme2tb/ffp/datasets/test/2019/perif/periphereies.shp',encoding='Windows-1253')
gdfperif=gdfperif.to_crs(4326)
```

```python
#df=pd.read_csv(r'/mnt/nvme2tb/ffp/datasets/test/2019/20190927_df.csv')
df=pd.read_csv(r'/mnt/nvme2tb/ffp/datasets/xai/20230825/20230825_xai_inp.csv',dtype={'id': str})
```

```python
df=extract_xy(df, xcoord='xlon', ycoord='ylat')
geom = gpd.points_from_xy(df['xlon'], df['ylat'])
gdf = gpd.GeoDataFrame(df, geometry=geom)
gdf=gdf.set_crs(4326)
```

```python
gdfp = gpd.sjoin(gdf, gdfperif.iloc[[9]], how='inner', predicate='within')
```
