```python
def xminmax(f81slice1, f81slice2, feat):
    resxmin=[]
    resxmax=[]
    for index1, row1 in f81slice1.iterrows():
        for index2, row2 in f81slice2.iterrows():
            xn1=row1['%s_x'%feat]
            xn2=row2['%s_x'%feat]
            x1=row1['%s_y'%feat]
            x2=row2['%s_y'%feat]
            if xn1==xn2:
                continue
            xmin=(xn1*x2-xn2*x1)/(xn1-xn2)
            resxmin.append(round(xmin*10**15))
            xmax=((x1-xmin+xn1*xmin)/xn1)
            resxmax.append(round(xmax*10**15))            
    return resxmin,resxmax

def filtminmax(mm1, mm2):
    xminl=set(mm1[0]).intersection(set(mm2[0]))
    xmaxl=set(mm1[1]).intersection(set(mm2[1]))
    return xminl,xmaxl
```

```python
min1,max1=xminmax(dfj.loc[dfj['f81_i']==72390], dfj.loc[dfj['f81_i']==187007],'x')
min2,max2=xminmax(dfj.loc[dfj['f81_i']==48260], dfj.loc[dfj['f81_i']==243619],'x')
filtminmax((min1,max1),(min2,max2))
```

```python
min1,max1=xminmax(dfj.loc[dfj['f81_i']==72390], dfj.loc[dfj['f81_i']==187007],'y')
min2,max2=xminmax(dfj.loc[dfj['f81_i']==48260], dfj.loc[dfj['f81_i']==243619],'y')
filtminmax((min1,max1),(min2,max2))
```
