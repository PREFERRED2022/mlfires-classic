```python
dfxai=pd.read_csv('/mnt/nvme2tb/ffp/datasets/xai/20190929_xai.csv', dtype={'id':str})
```

```python
dfxai['x']=dfxai['id'].str.slice(0,6).astype(int)/10000
```

```python
dfxai['y']=dfxai['id'].str.slice(6,12).astype(int)/10000
```

```python
dict0=dfxai.to_dict('records')[0]
dict0.pop('id')
dict0.pop('x')
dict0.pop('y')
```

```python

dict0
```

```python
def join_tuple_string(tup)->str:
    return '%s : %.2f'%(tup[0], tup[1])

numimp=5
rendict={c:'cor_%s'%(re.search('\d+',c).group(0)) for c in dfxai.columns if c.startswith('bin_corine')}
dfxai.rename(columns=rendict, inplace=True)
dictxai=dfxai.to_dict('records')
sortedimp=[]
for d in dictxai:
    d.pop('id')
    sortimp = sorted(d.items(), key=lambda x: x[1], reverse=True)
    sortedimp += [[sortimp[0][1], sortimp[1][1]]+list(map(join_tuple_string, sortimp[2:2+numimp]))]
dfxaifin=pd.DataFrame(sortedimp, columns=['y','x']+['imp%d'%i for i in list(range(1,numimp+1))])
    
```

```python
#print('Convert csv to shp points, input: %s, output: %s'%(inputfile, friskshp))
gdf = gpd.GeoDataFrame(dfxaifin, geometry=gpd.points_from_xy(dfxai['x'], dfxai['y'], z=None, crs=4326))
```

```python
gdf.drop(columns=['x', 'y'], inplace=True)
```

```python
gdf.to_file('/mnt/nvme2tb/ffp/datasets/xai/20190929_xai.shp')
```
