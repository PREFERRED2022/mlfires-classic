```python
def plot_histos(eradata,preddata,variables, sharey=True, flatarrays=False):
    
    n_bins = 100
    fig, axs = plt.subplots(1, 2*len(variables), sharey=sharey, tight_layout=True, figsize=(10*len(variables), 6))
    
    # We can set the number of bins with the *bins* keyword argument.
    varrng=', '.join([v['var'] for v in variables])+' '+eradata.time.min().item()+' to '+eradata.time.max().item()
    fig.suptitle(varrng)
    
    for i in range (0,len(variables)):
        xlim=variables[i]['xlim']
        eradataflat=eradata[variables[i]['var']].values.ravel() if not flatarrays else eradata[variables[i]['var']]
        axs[i*2].hist(eradataflat, bins=n_bins, range=xlim)
        axs[i*2].title.set_text('ERA5 mean: %.3f, std %.3f'%(eradataflat[~np.isnan(eradataflat)].mean(),
                                                             eradataflat[~np.isnan(eradataflat)].std()))
        pv0=preddata[variables[i]['var']].values.ravel() if not flatarrays else preddata[variables[i]['var']]
        pv0conv=eval(variables[i]['conv'])
        axs[i*2+1].hist(pv0conv, bins=n_bins, range=xlim)
        axs[i*2+1].title.set_text('Pred. mean: %.3f, std %.3f'%(pv0conv[~np.isnan(pv0conv)].mean(), 
                                                             pv0conv[~np.isnan(pv0conv)].std()))
        
    for i in range(2*len(variables)):
        axs[i].grid(which='both', axis='x')
    plt.show()
```

```python
def plot_histos_flat(eradata,preddata,convdata, variables, sharey=True):
    n_bins = 100
    fig, axs = plt.subplots(1, 3*len(variables), sharey=sharey, tight_layout=True, figsize=(10*len(variables), 6))
    
    # We can set the number of bins with the *bins* keyword argument.
    #varrng=', '.join([v['var'] for v in variables])+' '+eradata.time.min().item()+' to '+eradata.time.max().item()
    fig.suptitle('converted comparison')
    
    for i in range (0,len(variables)):
        xlim=variables[i]['xlim']
        axs[i*2].hist(eradata, bins=n_bins, range=xlim)
        axs[i*2].title.set_text('ERA5 mean: %.3f, std %.3f'%(eradata[~np.isnan(eradata)].mean(),
                                                             eradata[~np.isnan(eradata)].std()))
        axs[i*2+1].hist(preddata, bins=n_bins, range=xlim)
        axs[i*2+1].title.set_text('Pred. mean: %.3f, std %.3f'%(preddata[~np.isnan(preddata)].mean(), 
                                                             preddata[~np.isnan(preddata)].std()))
        axs[i*2+2].hist(convdata, bins=n_bins, range=xlim)
        axs[i*2+2].title.set_text('Pred. Converted mean: %.3f, std %.3f'%(convdata[~np.isnan(convdata)].mean(), 
                                                             convdata[~np.isnan(convdata)].std()))
        
    for i in range(3*len(variables)):
        axs[i].grid(which='both')
    plt.show()
```

```python
def plot_scatter(eradata,preddata,variables, sharey=True):
    fig, axs = plt.subplots(1, len(variables), sharey=sharey, tight_layout=True, 
                            figsize=(10*len(variables), 6),squeeze=False)
    varrng=', '.join([v['var'] for v in variables])+' '+eradata.time.min().item()+' to '+eradata.time.max().item()
    fig.suptitle(varrng)
    
    for i in range (0,len(variables)):
        pv0=preddata[variables[i]['var']].values.ravel()
        pv0conv=eval(variables[i]['conv'])
        axs[0][i].scatter(eradata[variables[i]['var']].values.ravel(), pv0conv)
    for i in range(len(variables)):
        axs[0][i].grid(which='both')
        axs[0][i].set_xlabel('ERA5')
        axs[0][i].set_ylabel('Prediction')

    plt.show()
```

```python
def load_day(day):
    folderERA5='/mnt/nvme2tb/ffp/datasets/test/2021'
    monthpred=None
    monthERA5=None
    f=os.path.join('/mnt/nvme2tb/ffp/datasets/test/bartsotas2021',day+'.nc')
    fERA5=validfile(f, folderERA5)
    if not fERA5:
        print('Invalid file %s'%f)
    else:
        monthpred, monthERA5=concatfiles(monthpred, monthERA5,f,folderERA5)   
    return monthERA5, monthpred    
    
def load_month(month):
    setmonth = [f for f in fileutils.find_files('/mnt/nvme2tb/ffp/datasets/test/bartsotas2021', '*2021%s*'%month, listtype="list")]
    folderERA5='/mnt/nvme2tb/ffp/datasets/test/2021'
    monthpred=None
    monthERA5=None
    for f in tqdm(setmonth, ncols=80):
        monthpred, monthERA5=concatfiles(monthpred, monthERA5,f,folderERA5)   
    return monthERA5, monthpred
```

```python
def validfile(f, folderERA5):
    #bname=os.path.basename(f)
    era5fname=re.search('^(.*?)\.nc',os.path.basename(f)).group(1)
    fera5=os.path.join(folderERA5,era5fname+'_df.nc')
    if os.path.isfile(fera5):
        return fera5
    else:
        return False

def applygreecemask(dayxr,grmask):
    grmask1int=grmask.squeeze().interp(y=dayxr["y"], x=dayxr["x"], method="nearest")
    grappliedmask=dayxr*grmask1int
    return grappliedmask
    
def concatfiles(monthpred, monthERA5,fpred, folderERA5, applygrmask=True):
    grmask=rxr.open_rasterio('/mnt/nvme2tb/ffp/datasets/protipo_cropped_greece.tif')
    grmask1=grmask.squeeze().where(grmask.isnull(), 1)
    fERA5=validfile(fpred, folderERA5)
    if fERA5:
        dayb=xr.open_dataset(fpred)
        if applygrmask:
            dayb=applygreecemask(dayb,grmask1)
        monthpred=xr.concat([monthpred,dayb],"time") if monthpred is not None else dayb
        day=xr.open_dataset(os.path.join(folderERA5,fERA5))
        if applygrmask:
            day=applygreecemask(day,grmask1)
        #day=day.expand_dims(dim={'time':1})
        day=day.assign_coords({"time": dayb.time.values})
        monthERA5=xr.concat([monthERA5,day],"time") if monthERA5 is not None else day
    return monthpred, monthERA5
```
